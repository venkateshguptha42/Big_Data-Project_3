```python
from kafka import KafkaConsumer
import matplotlib.pyplot as plt
import json
import threading
```

```python
consumer = KafkaConsumer('tata1',
                         group_id='tata1',
                         bootstrap_servers=['localhost:9092'],
                        )
print("consumer started ...")
```

```python
x = {}
def plot():
    global x
    for message in consumer:
        x[json.loads((message.value).decode("utf-8"))["date"]] = (json.loads((message.value).decode("utf-8"))["volume"])
print(x)
```

```python
plot_thread = threading.Thread(target=plot)
plot_thread.start()
```

```python
fig = plt.figure(figsize = (12, 6))
x = dict(sorted(x.items(), key=lambda item: item[0], reverse=False))
plt.bar([*x.keys()][:30], [*x.values()][:30], width=0.1)
plt.xlabel("Date", fontsize=15)
plt.ylabel("Volume", fontsize=15)
plt.xticks(rotation=90)
plt.title("TATA MOTORS SHARES", fontsize=15)
```
